### P&L Kuralları

#### Revenues  (Gelirler )
**Sales fees:** Olist, teslim edilen her order’ın ürün fiyatı üzerinden (freight hariç) **%10 pay** alır.  
**Subscription fees:** Olist, seller başına ayda **80 BRL** ücret alır.

#### Costs (Maliyetler)
Kötü review’lara (<= 3 stars) sahip orders için _tahmini_ **reputation costs**  

💡 Uzun vadede kötü customer experience iş açısından sonuçlara yol açar: düşük tekrar satın alma oranı, anında customer support maliyeti, refunds veya olumsuz word of mouth iletişimi. Her kötü review için parasal bir cost varsayımı yapıyoruz:

```python
# review_score: cost(BRL)
{'1 star': 100
'2 stars': 50
'3 stars': 40
'4 stars': 0
'5 stars': 0}
```

**IT costs:** Olist'in total cumulated IT Costs değeri, platforma şimdiye kadar katılmış toplam seller sayısının karekökü ile, şimdiye kadar satılmış toplam birikimli item sayısının karekökü oranında ölçeklenir.

$IT\_costs = \alpha * \sqrt{n\_sellers} + \beta * \sqrt{n\_items}$  

Olist'in data team’i bize bu scaling parameter’ları için şu değerleri verdi:
- $\alpha = 3157.27$
- $\beta = 978.23$

💡 Yönetilen seller sayısı da, sales transaction sayısı da IT systems için maliyetlidir.
💡 Ancak karekök kullanımı scale-effects’i gösterir: IT-system’ler büyüdükçe daha verimli olma eğilimindedir.
💡 Alpha > Beta olması, Olist’in çok item satan az sayıda seller ile çalıştığında, bunun tersi senaryoya göre daha düşük IT Cost’a sahip olduğu anlamına gelir.

1000 sellers ve toplam 100 items satışıyla toplam IT cost 109,624 BRL olur

100 sellers ve toplam 1000 items satışıyla toplam IT cost 62,507 BRL olur

Son olarak, IT department size marketplace’in kuruluşundan bu yana birikmiş IT costs toplamının 500,000 BRL olduğunu da iletti.

Şimdi, bu IT cost formülünü biraz daha iyi anlamak için bir simulation çalıştıralım. Kod üzerinde zaman harcamanıza gerek yok, sadece ortaya çıkan chart’lara bakın.

````python
%load_ext autoreload
%autoreload 2

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

alpha = 3157.27
beta = 978.23

x_sellers = np.arange(3_000)
y_it_costs_sellers = alpha * x_sellers ** 0.5

x_items = np.arange(100_000)
y_it_costs_items = beta * x_items ** 0.5

plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.plot(x_sellers, y_it_costs_sellers, label='IT costs per seller', color='blue')
plt.title('IT Costs vs. Number of Sellers')
plt.subplot(1, 2, 2)
plt.plot(x_items, y_it_costs_items, label='IT costs per item', color='orange')
plt.title('IT Costs vs. Number of Items')
plt.suptitle('IT Costs Simulation');
````

### 2.2 IT Maliyetleri Hariç Varsayım Analizi

👉 Gerçek analizi yapma zamanı. İlk olarak bunu IT costs **dahil etmeden** gerçekleştireceğiz.

Adımlarımız:  

1️⃣ Sellers verisini load edin ve profits değerine göre azalan şekilde sıralayın.

2️⃣ Cumulative değerleri hesaplayın:
   - Her satır için cumulative revenues hesaplayın.
   - Her satır için cumulative profits hesaplayın.
   - Her satır için cumulative items sold hesaplayın.

   Pandas ile cumulative sum nasıl hesaplanır, araştırın.

3️⃣ Sonuçlarınızı plot edin ve analiz edin. Pandas’ın `.plot()` method’unu kullanın.

4️⃣ Profits’e göre tutulması gereken optimum seller sayısını belirleyin. (İpucu: `np.argmax` araştırın). Bu kararın etkisi ne olurdu:
   - Net profit (IT costs dahil etmediğimiz için IT costs öncesi)?

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 1. Satıcıları Net Katkılarına (Revenue - Review Cost) göre EN İYİDEN EN KÖTÜYE sırala
# Bu çok kritik; kârı maksimize eden noktayı görmek için en değerliden başlamalıyız.
sellers_analysis_sorted = seller_analysis.sort_values(by='net_contribution', ascending=False).reset_index(drop=True)

# 2. Kümülatif Hesaplamalar (Adım adım her satıcıyı eklediğimizde toplam ne oluyor?)
sellers_analysis_sorted['cum_profits'] = sellers_analysis_sorted['net_contribution'].cumsum()
sellers_analysis_sorted['n_sellers'] = sellers_analysis_sorted.index + 1

# 3. Optimum Noktayı Bul (IT Hariç)
# Kârın en yüksek olduğu satır, aslında zarar ettirenlerin başladığı yerin hemen öncesidir.
best_idx = sellers_analysis_sorted['cum_profits'].idxmax()
optimum_n = sellers_analysis_sorted.loc[best_idx, 'n_sellers']
max_profit = sellers_analysis_sorted.loc[best_idx, 'cum_profits']

# 4. Görselleştirme (CEO Sunum Kalitesinde)
plt.figure(figsize=(12, 6))
plt.plot(sellers_analysis_sorted['n_sellers'], sellers_analysis_sorted['cum_profits'],
         color='#1f77b4', linewidth=3, label='Cumulative Net Contribution')

# Zirve noktasını işaretle
plt.scatter(optimum_n, max_profit, color='red', s=100, zorder=5)
plt.axvline(optimum_n, color='red', linestyle='--', alpha=0.6,
            label=f'Optimum Point ({optimum_n} Sellers)')

# Grafik Süslemeleri
plt.title("Olist Profit Optimization (Excluding IT Costs)", fontsize=16, pad=20)
plt.xlabel("Number of Sellers (Ranked by Profitability)", fontsize=12)
plt.ylabel("Total Profit (BRL)", fontsize=12)
plt.grid(True, which='both', linestyle='--', alpha=0.5)
plt.legend(fontsize=11)

# Şık bir arka plan rengi
plt.gca().set_facecolor('#f9f9f9')
plt.show()

print(f"📊 CEO ÖZETİ (IT Hariç):")
print(f"--------------------------------------------------")
print(f"Toplam Satıcı Sayısı: {len(sellers_analysis_sorted)}")
print(f"Tutulması Gereken (Kârlı) Satıcı Sayısı: {optimum_n}")
print(f"Sistemden Çıkarılması Gereken Zararlı Satıcı Sayısı: {len(sellers_analysis_sorted) - optimum_n}")
print(f"Maksimum Kâr Potansiyeli: {max_profit:,.2f} BRL")
print(f"--------------------------------------------------")
# 4. Optimum Noktayı Bul (IT maliyeti yokken kârın zirve yaptığı yer)
best_n_it_hariç = sellers_sorted['cum_profits'].idxmax() + 1
max_profit_it_hariç = sellers_sorted['cum_profits'].max()

print(f"👉 IT Maliyetleri Hariç Sonuç:")
print(f"En iyi {best_n_it_hariç} satıcıyı tutmalı, geri kalanları çıkarmalıyız.")
print(f"Bu senaryoda toplam kâr: {max_profit_it_hariç:,.2f} BRL")
````

### 2.3 IT maliyetleri ile varsayımsal analiz (İsteğe bağlı)

👉 Bu kez IT costs dahil edilerek gerçek analizi yapma zamanı.

Adımlarımız şunlar:  

1️⃣ İki parametreye göre IT costs hesaplayan bir function yazın: seller sayısı ve item sayısı. Temelde matematiksel formülü Python kullanarak yazmanız yeterli.

2️⃣ Sellers verisini load edin ve (IT costs öncesi) kârlarına göre azalan şekilde sıralayın.

3️⃣ Cumulative değerleri hesaplayın (önceki kodu yeniden kullanın):
   - Her satır için cumulative revenues hesaplayın.
   - Her satır için cumulative profits hesaplayın.
   - Her satır için toplam satılan item miktarının cumulative değerini hesaplayın.  

4️⃣ IT costs etkisini hesaplayın:
   - 1️⃣. adımda yazdığınız function’ı kullanarak her satır için cumulative IT cost ekleyin.
   - IT costs sonrası net profit’i hesaplayın.

5️⃣ Sonuçlarınızı plot edin ve analiz edin.

6️⃣ Bu kez IT costs sonrası kârları baz alarak tutulması gereken optimum seller sayısını belirleyin. Bu kararın etkisi ne olurdu:
   - IT costs sonrası net profit?
   - IT costs öncesi net profit?
   - IT costs?
   - Total revenues?
   - Seller sayısı?
   - Satılan item sayısı?

Analizinizde IT costs ne kadar önemliydi?


````python
# 1. IT Maliyet Fonksiyonunu Tanımla (Parametreleri yukarıda tanımladığını varsayıyorum)
def calculate_it_cost(n_sellers, n_items):
    return ALPHA * (n_sellers**0.5) + BETA * (n_items**0.5)

# 2. Veriyi En İyi Satıcıdan En Kötüye Sırala (Zaten sıralıysa sorun yok)
sellers_final = seller_analysis.sort_values(by='net_contribution', ascending=False).reset_index(drop=True)

# 3. Kümülatif Değerleri Hesapla
sellers_final['cum_net_contribution'] = sellers_final['net_contribution'].cumsum()
sellers_final['cum_n_items'] = sellers_final['n_items'].cumsum()
sellers_final['n_sellers'] = sellers_final.index + 1

# 4. Her Satır İçin Kümülatif IT Maliyetini ve Final Kârı Hesapla
sellers_final['cum_it_cost'] = sellers_final.apply(
    lambda row: calculate_it_cost(row['n_sellers'], row['cum_n_items']), axis=1
)
sellers_final['final_net_profit'] = sellers_final['cum_net_contribution'] - sellers_final['cum_it_cost']

# 5. Gerçek Optimum Noktayı Bul (IT Dahil)
best_idx_final = sellers_final['final_net_profit'].idxmax()
optimum_n_final = sellers_final.loc[best_idx_final, 'n_sellers']
max_profit_final = sellers_final.loc[best_idx_final, 'final_net_profit']

# 6. Görselleştirme (Profesyonel Seviye)
plt.figure(figsize=(12, 6))

# IT Hariç Kar (Karşılaştırma için silik çizgi)
plt.plot(sellers_final['n_sellers'], sellers_final['cum_net_contribution'],
         color='gray', linestyle='--', alpha=0.5, label='Profit (Excl. IT Costs)')

# IT Dahil Gerçek Kar
plt.plot(sellers_final['n_sellers'], sellers_final['final_net_profit'],
         color='#d35400', linewidth=3, label='Final Net Profit (Incl. IT Costs)')

# Zirveyi İşaretle
plt.scatter(optimum_n_final, max_profit_final, color='black', s=100, zorder=5)
plt.axvline(optimum_n_final, color='black', linestyle=':',
            label=f'True Optimum: {optimum_n_final} Sellers')

plt.title("Olist Profit Optimization: THE FINAL TRUTH", fontsize=16, pad=20)
plt.xlabel("Number of Sellers (Best to Worst)", fontsize=12)
plt.ylabel("BRL", fontsize=12)
plt.legend()
plt.grid(True, alpha=0.2)
plt.show()

print(f"🚀 FINAL KARAR PARAMETRELERİ:")
print(f"--------------------------------------------------")
print(f"IT Öncesi Optimum Satıcı Sayısı: {optimum_n} (Sadece kendi kârı)")
print(f"IT DAHİL Optimum Satıcı Sayısı: {optimum_n_final}")
print(f"FARK: {optimum_n - optimum_n_final} satıcı daha 'sistem yükü' nedeniyle elenmeli.")
print(f"Maksimum Net Kar: {max_profit_final:,.2f} BRL")
print(f"--------------------------------------------------")
````
